### MODELANDO

```python
datafile='df_BACH__manerauno.csv'
data = np.genfromtxt( #csv to array
    datafile,
    delimiter=",",
    usecols=range(9, 24),#tomar el array a partir de la columna 8, a partir de days_left
    skip_header=1
)
data
```

```python
data=np.nan_to_num(data)
```

### constyendo pipeline

```python
preprocessor = Pipeline(
    [
        ("scaler", MinMaxScaler()),
        ("pca", PCA(n_components=1, random_state=42)), #pasamos solo 2 componentes para queluego se veaa bien con graficos y se pueda representar
    ]
)
```

```python
clusterer = Pipeline(
   [
       (
           "kmeans",
           KMeans(
               n_clusters=23,
               init="k-means++",
               n_init=50,
               max_iter=500,
               random_state=42,
           ),
       ),
   ]
)
```

```python
pipe = Pipeline(
    [
        ("preprocessor", preprocessor),
        ("clusterer", clusterer)
    ]
)
```

```python
pipe.fit(data)
```

### con manerauno

```python
preprocessed_data = pipe["preprocessor"].transform(data)

predicted_labels = pipe["clusterer"]["kmeans"].labels_

silhouette_score(preprocessed_data, predicted_labels)#if pca k=3
```

```python
pipe["clusterer"]["kmeans"].inertia_
```

```python
# Empty lists to hold evaluation metrics
silhouette_scores = []
sse = []
ari_scores = []
for n in range(6, 40):
    # This set the number of components for pca,
    # but leaves other steps unchanged
    pipe["preprocessor"]["pca"].n_components = 1
    pipe["clusterer"]["kmeans"].n_clusters = n
    pipe.fit(data)

    silhouette_coef = silhouette_score(
        pipe["preprocessor"].transform(data),
        pipe["clusterer"]["kmeans"].labels_,
    )
    
    sse.append(pipe["clusterer"]["kmeans"].inertia_)
    # Add metrics to their lists
    silhouette_scores.append(silhouette_coef)
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5.5, 4))
plt.plot(
    range(6, 40),
    sse,
    c="#008fd5",
    label="Silhouette Coefficient",
)

plt.xlabel("n_means")
plt.legend()
plt.title("Clustering Performance as a Function of n_components")
plt.tight_layout()
plt.show()
```

```python
kl = KneeLocator(
    range(6, 40), sse, curve="convex", direction="decreasing"
)

kl.elbow
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(5.5, 4))
plt.plot(
    range(6, 40),
    silhouette_scores,
    c="#008fd5",
    label="Silhouette Coefficient",
)

plt.xlabel("n_means")
plt.legend()
plt.title("Clustering Performance as a Function of n_components")
plt.tight_layout()
plt.show()
```

```python
nueveKmeansSCORE=[]
sseSCORE=[]
ks=[]
pcaks=[]
for i2 in range(6,24):

    for i in range(1,3):
        preprocessor = Pipeline(    
            [
                ("scaler", MinMaxScaler()),
                ("pca", PCA(n_components=i, random_state=42)), #pasamos solo 2 componentes para queluego se veaa bien con graficos y se pueda representar
            ]
        )
        clusterer = Pipeline(
           [
               (
                   "kmeans",
                   KMeans(
                       n_clusters=i2,
                       init="k-means++",
                       n_init=50,
                       max_iter=500,
                       random_state=42,
                   ),
               ),
           ]
        )
        pipe = Pipeline(
            [
                ("preprocessor", preprocessor),
                ("clusterer", clusterer)
            ]
        )
        pipe.fit(data)
        preprocessed_data = pipe["preprocessor"].transform(data)

        predicted_labels = pipe["clusterer"]["kmeans"].labels_
        sse = pipe["clusterer"]["kmeans"].inertia_
        nueveKmeansSCORE.append(silhouette_score(preprocessed_data, predicted_labels))
        sseSCORE.append(sse)
        ks.append(i2)
        pcaks.append((i))
        print("k="+str(i2)+" resultado(SSE) de realizar el pca con "+str(i)+" componentes",sse)
        print("k="+str(i2)+" resultado(SILUETE) de realizar el pca con "+str(i)+" componentes",silhouette_score(preprocessed_data, predicted_labels))
```
