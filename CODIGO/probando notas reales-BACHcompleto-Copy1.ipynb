### Resumen CIENCIAS_TECNOLOGICAS

```python
#df_BACH__SOCIALES_CIENCIAS.index=(range(1,6001))
fails=[]
dft=df_BACH__SOCIALES_CIENCIAS.iloc[list(range(0,2000))]
for i,pos2 in enumerate(dft[["Lengua","Mates","Ingles","Fisica","Dibujo","Historia"]]):
    suspensos=[]
    suspensos2=[]
    sobresalientes=[]
    fails1=[]
    for i,pos in enumerate(dft[pos2].values.tolist()):
        fails1.append(pos)
        if pos <5:
            
            if pos >= 4:
                suspensos.append(pos)
            else:
                suspensos2.append(pos)
        elif pos >8:
            sobresalientes.append(pos)

    print(round(Average(dft[pos2].values.tolist()),2),"media de "+pos2)
    print(len(suspensos2),"alumnos con menos de un 4")
    fails.append(len(fails1))
    print(len(suspensos2)/20+(len(suspensos)/20),"% de alumnos con notas menores a 5")
    print(len(suspensos2)/20,"% de alumnos con notas menores a 4")
    print(len(sobresalientes)/20,"% de alumnos con notas mayores a 8")
    print()
```
